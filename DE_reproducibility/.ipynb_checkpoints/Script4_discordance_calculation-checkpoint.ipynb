```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

from tqdm import tqdm
from tqdm import trange

import random
import itertools

n, n_iter, alpha, lfc_min_thershold, base_mean_min_threshold, min_percentage, max_cutoff, normalization_type = 1000, 2, 0.1, 0.1, 0.1, 0.0, 0.2, 'RUV'

import statsmodels.api as sm
from scipy.stats import mannwhitneyu
from decimal import Decimal
```

```python
subset_size = 0.01
```

```python
def flatten_list(list1):
    return [item for sublist in list1 for item in sublist]
```

```python
p = open("../data/actual_proportions.n."+ str(n) + '.n_iter.' + str(n_iter) + '.txt').read().splitlines()
p = sorted(set([float(i) for i in p]))
```

```python
concordance_genes = open('../data/' + normalization_type + 'concordance_genes_n_' + str(n) + '_n_iter_' + str(n_iter) + '.txt').read().splitlines()
lfc_df = pd.read_csv('../data/' + normalization_type + 'parsed_lfc_df_n' + str(n) + '_n_iter_' + str(n_iter) + '.csv', index_col = 0)

base_mean_df = pd.read_csv('../data/' + normalization_type + 'parsed_base_mean_df_n' + str(n) + '_n_iter_' + str(n_iter) + '.csv', index_col = 0)
# log psudeocount
base_mean_df = np.log(np.add(base_mean_df,10**-6))

lfc_SE_df = pd.read_csv('../data/' + normalization_type + 'parsed_lfc_SE_df_n' + str(n) + '_n_iter_' + str(n_iter) + '.csv', index_col = 0)
lfc_dispersion_df = pd.read_csv('../data/' + normalization_type + 'parsed_lfc_dispersion_df_n' + str(n) + '_n_iter_' + str(n_iter) + '.csv', 
                                index_col = 0)
```

```python
print('Stratify concordance by freeze thaw and RIN')

print('Load colLabel to subset dict')
colLabel_to_subset_ = pd.read_csv('../data/colLabel_to_subset_n_' + str(n) + '_n_iter_' + str(n_iter) + '.csv', 
                                 index_col = 0)
colLabel_to_subset = dict(zip(colLabel_to_subset_['colLabel'].tolist(),colLabel_to_subset_['subset'].tolist()))
colLabel_to_subset_.set_index('colLabel', inplace = True, drop = True)
annot = pd.read_excel('../data/manifest.xlsx',sheet_name = 5)

print('Get average freeze thaw and RIN per subset') # independent subsets, not subset combinations
ft_list = [annot['final_freeze_thaw'][annot['Comment'].isin(colLabel_to_subset[i].split(' '))].mean() for i in colLabel_to_subset_.index]
RIN_list = [annot['RINe'][annot['Comment'].isin(colLabel_to_subset[i].split(' '))].mean() for i in colLabel_to_subset_.index]
summary = colLabel_to_subset_.copy()
summary['Avg Freeze Thaw'] = ft_list
summary['Avg RIN'] = RIN_list
summary.rename(columns = {'subset': 'Subset'}, inplace = True)
```

```python
def subset_size(x):
    return len(x.split(' '))
summary['Subset Size'] = summary['Subset'].apply(subset_size)
summary['Proportion'] = summary['Subset Size'].map(dict(zip(sorted(summary['Subset Size'].unique()),p)))
```

```python
for effect_size_metric in ['ALT', 'COHEND', 'MEAN']:
    summary_df = summary.copy()
    metrics = ['Avg Freeze Thaw', 'Avg RIN']
    splits = [2,2]

    for i in range(len(metrics)):
        metric, split = metrics[i], splits[i]
        def bin_by_proportion(x):
            try:
                return pd.qcut(x,split)
            except:
                print('Quantiles did not work for ' + metric + ' by Subset Size')
                return pd.cut(x, split)
        summary_df[metric + ' Bins by Proportion'] = summary_df.groupby('Subset Size')[metric].apply(bin_by_proportion)

    summary_df.sort_values(by = 'Subset Size', inplace = True)

    metrics = ['Avg Freeze Thaw', 'Avg RIN']
    for i in range(len(metrics)):
        metric = metrics[i]
        test_index = []
        bin_rank_list = []
        for p_ in sorted(set(summary_df['Subset Size'])):

            d = summary_df[summary_df['Subset Size'] == p_]
            test_index.append(d.index.tolist())
            bin_rank = dict(zip(sorted(d[metric + ' Bins by Proportion'].unique()), [0,1]))
            bin_rank_list += d[metric + ' Bins by Proportion'].map(bin_rank).tolist()

        if flatten_list(test_index) == summary_df.index.tolist():
            summary_df[metric + ' Ranks by Proportion'] = bin_rank_list
        else:
            raise ValueError('Did not work')


    freeze_thaw_map = {0: 'High Sample Quality', 1: 'Low Sample Quality'}
    RIN_map = {1: 'High Sample Quality', 0: 'Low Sample Quality'}
    maps = [freeze_thaw_map, RIN_map]

    for i in range(len(metrics)):
        metric = metrics[i]
        summary_df[metric + ' Ranks by Proportion'] = summary_df[metric + ' Ranks by Proportion'].map(maps[i])


    string = '../data/' + normalization_type + 'summary_metrics_each_subset'
    string += '_n_' + str(n) + '_n_iter_' + str(n_iter) + '.csv'
    summary_df.to_csv(string)

    print('Generate concordance df by proportion and by bin rank')
    cols_dict = {}
    counter = 0
    for i in range(len(metrics)):
        metric = metrics[i]
        split = splits[i]
        for p_ in p:
            for k in ['High Sample Quality', 'Low Sample Quality']:
                for stat in ['SD', 'Mean']:
                    cols_dict[counter] = {}
                    cols_dict[counter]['Proportion'] = p_
                    cols_dict[counter]['Metric'] = metric
                    cols_dict[counter]['Bin Rank'] = k
                    cols_dict[counter]['Statistic'] = stat
                    counter += 1

    concordance_df = pd.DataFrame(columns = cols_dict.keys(), index = concordance_genes) 

    for col in concordance_df.columns:
        p_ = cols_dict[col]['Proportion']
        m = cols_dict[col]['Metric']
        r = cols_dict[col]['Bin Rank']
        stat = cols_dict[col]['Statistic']
        index = summary_df[(summary_df['Proportion'] == p_) & (summary_df[m + ' Ranks by Proportion'] == r)].index.tolist()
        if stat == 'SD':
            d = lfc_df.loc[concordance_genes,index]
            val = d.std(axis = 1).tolist()
        elif stat == 'Mean':
            _lfc = lfc_df.loc[concordance_genes,index].median(axis = 1).abs().values
            _basemean = base_mean_df.loc[concordance_genes,index].median(axis = 1).values
            _lfcSE = lfc_SE_df.loc[concordance_genes,index].median(axis = 1).abs().values
            _lfcdisp = lfc_dispersion_df.loc[concordance_genes,index].median(axis = 1).abs().values

            if effect_size_metric == 'MEAN':
                val = lfc_df.loc[concordance_genes,index].median(axis = 1).abs().tolist()

            # Cohen's d
            elif effect_size_metric == 'COHEND':
                val = np.divide(_lfc, np.sqrt(_lfcSE))
            elif effect_size_metric == 'ALT':
                # Love effect size: LFC/[sqrt( (1/basemean) + dispersion)]
                # https://www.biostars.org/p/140976/&sa=D&ust=1569362924764000&usg=AFQjCNHFX1v1USe7Gizo5PI95XKLqjqugA
                val = np.divide(_lfc,np.sqrt(np.add(np.divide(1,_basemean), _lfcdisp)))
            else:
                raise ValueError('Must choose an appropriate effect size metric')



        concordance_df.loc[:,col] = val    

    a = sorted(set(summary_df['Proportion']))
    b = [4,6,8,10,12,14]
    proportion_to_size = dict(zip(a,b))
    
    string = '../data/' + normalization_type + 'concordance_df'
    string += '_effectsizemetric_' + effect_size_metric + '_n_' + str(n) + '_n_iter_' + str(n_iter) + '.csv'
    concordance_df.to_csv(string)
    
    def generate_df_for_concordance_stats(concordance_df, cols_dict, proportion_to_size):
        '''Function bc  generating difference concordance dataframes. Formats to run stats in R.'''
        n_genes = concordance_df.shape[0]
        c_stats = pd.DataFrame(index = range(n_genes *int(concordance_df.shape[1]*0.5)), 
                               columns = (['Discordance', 'Effect Size'] + list(cols_dict[0].keys())[:-1]))

        n_shape = 0
        mD = list(cols_dict[0].keys())[:-1]

        for i in cols_dict.keys():
            if cols_dict[i]['Statistic'] == 'SD':
                c_stats.loc[n_shape:(n_shape + n_genes)-1,'Discordance'] = concordance_df.loc[:,i].tolist()
            elif cols_dict[i]['Statistic'] == 'Mean':
                c_stats.loc[n_shape:(n_shape + n_genes)-1,'Effect Size'] = concordance_df.loc[:,i].tolist()
            if (i % 2) == 1: # only needs to be done once
                for md in mD: # add metadata
                    c_stats.loc[n_shape:(n_shape + n_genes)-1,md] = [cols_dict[i][md]]*n_genes
                if (i < list(cols_dict.keys())[-1]-1): # maintain appropriate index for c_stats
                    n_shape = n_shape + n_genes
        c_stats['Discordance'] = c_stats['Discordance'].astype('float')
        c_stats['Effect Size'] = c_stats['Effect Size'].astype('float')

        # label formatting
        if 'Bin Rank' in c_stats.columns: 
            c_stats['Bin Rank'] = c_stats['Bin Rank'].map({'High Sample Quality' : 1, 'Low Sample Quality': 0})
            c_stats.rename(columns = {'Bin Rank': 'Sample Quality'}, inplace= True)

        if 'Proportion' in c_stats.columns:        
            c_stats['Proportion'] = c_stats['Proportion'].map(proportion_to_size) # subset size instead
            c_stats.rename(columns = {'Proportion': 'Subset Size'}, inplace = True)
        # error check
        if c_stats.isna().values.any():
            raise ValueError('Something went wrong when generating concordance stats df')
        return c_stats

    c_stats = generate_df_for_concordance_stats(concordance_df, cols_dict, proportion_to_size)
    string = '../data/' + normalization_type + 'concordance_df_forGLM_sizeandquality'
    string += '_effectsizemetric_' + effect_size_metric + '_n_' + str(n) + '_n_iter_' + str(n_iter) + '.csv'
    c_stats.to_csv(string)
```

```python
print('Concordance by proportion')
sns.set(style = 'white', font_scale = 2)
fig, ax = plt.subplots(figsize = (10,10))
counter = 0
sns.set(style = 'white', font_scale = 2)
colors = sns.color_palette("Blues")

# stats df
concordance_stats = pd.DataFrame(index = ['Coefficient', 'Coefficient P value', 'Intercept', 
                                      'Adjusted Coefficient of Determination'])

max_x = []
for p_ in tqdm(sorted(set(summary_df['Proportion']))):
    d = summary_df[summary_df['Proportion'] == p_]

    X_col = [k for k in cols_dict.keys() if (cols_dict[k]['Proportion'] == p_)
             & (cols_dict[k]['Statistic'] == 'Mean')][0]


    Y_col = [k for k in cols_dict.keys() if (cols_dict[k]['Proportion'] == p_)
             & (cols_dict[k]['Statistic'] == 'SD')][0]


    if type(X_col) == list:
        X = concordance_df[X_col].mean(axis = 1)
    else:
        X = concordance_df[X_col]

    Y = concordance_df[Y_col]

    max_x.append(X.max())

    sns.regplot(x = X, y = Y, scatter = False, color = colors[counter], truncate = True,
               ax = ax)
    counter += 1


#-----------### STATS --------------------
    X = sm.add_constant(X)
    reg = sm.OLS(Y, X).fit().summary()

    stats = pd.read_html(reg.tables[1].as_html(), header = 0, index_col=0)[0]
    coeff = stats.iloc[1,0]
    pval = stats.iloc[1,3]
    intercept = stats.loc['const', 'coef']
    adj_R_squared = pd.read_html(reg.tables[0].as_html(), index_col=0)[0].iloc[1,2]
    vals = [coeff, pval, intercept, adj_R_squared]

    concordance_stats[Y_col] = vals
#-----------### END STATS --------------------


    labels = [proportion_to_size[p_] for p_ in sorted(set(summary_df['Proportion']))]
    ax.legend(labels = labels, fontsize = 15, title = 'Subset Size')
    ax.set_xlim([0,max(max_x)])
    ax.set_ylabel('Discordance')
    if effect_size_metric == 'ALT':
        ax.set_xlabel("Mean-Variance Standardized Effect")
    else:
        ax.set_xlabel("Effect Size")


```
