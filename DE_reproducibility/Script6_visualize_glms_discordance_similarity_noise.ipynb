```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd
import numpy as np
import statsmodels
from statsmodels.stats import multitest as MT
n, n_iter, alpha, lfc_min_thershold, base_mean_min_threshold, min_percentage, max_cutoff, normalization_type = 1000, 2, 0.1, 0.1, 0.1, 0.0, 0.2, 'RUV'
effect_size_metric = 'ALT'
```

Apply BH correction to Discordance GLMs

```python
string = '../data/supplementary_discordance_GLM_2'
string += '_n_' + str(n) +'_n_iter_' + str(n_iter) +'.csv'

dg = pd.read_csv(string, index_col = 0) 
# filter 
dg = dg[(dg['nt'] == normalization_type)]
dg.drop(columns = ['nt'], inplace = True)
dg.rename(columns = {'Pr...t..': 'p_value'}, inplace = True)
dg['Benjamini-Hochberg FDR corrected p-value'] = [float('nan')]*dg.shape[0]

for esm in dg['esm'].unique():
    index = dg[dg['esm'] == esm].index
    bh = MT.multipletests(dg.loc[index, 'p_value'], alpha=0.05, method='fdr_bh')
    dg.loc[index, 'Benjamini-Hochberg FDR corrected p-value'] = bh[1]

map_ = {'ALT': 'mean-variance standardized effect', 'COHEND': "Cohen's d", 'MEAN': 'Median'}
dg['esm'] = dg['esm'].map(map_)
dg.rename(columns={'esm': 'Effect Size Metric'}, inplace = True)
string = '../data/supplementary_discordance_GLM_withBH_2_n_1000_n_iter_2.csv'
dg.to_csv(string)
```

# Discordance

GLM(Discordance ~ Effect Size * Sample Quality)

```python
string = '../data/supplementary_discordance_GLM_2'
string += '_n_' + str(n) +'_n_iter_' + str(n_iter) +'.csv'

dg = pd.read_csv(string, index_col = 0) 
# filter 
dg = dg[(dg['nt'] == normalization_type) 
        & (dg['esm'] == effect_size_metric)].iloc[:,2:]


# format
dg.reset_index(inplace = True, drop = True)

# apply a benjamini test correction
bh = statsmodels.stats.multitest.multipletests(dg['Pr...t..'], alpha=0.05, method='fdr_bh')
dg['Pr...t..'] = bh[1]

keys = dg['GLM.Parameter'].unique().tolist()
values = [' '.join(i.split('.')) for i in keys]
values[values.index('(Intercept)')] = 'Intercept'
format_map = dict(zip(keys,values))

dg['GLM.Parameter'] = dg['GLM.Parameter'].map(format_map)
dg.rename(columns = {dg.columns[-1]: 'p-value'}, inplace = True)

def binarize(x):
    if x > 0.1:
        return 0
    else:
        return 1
    
dg['p-value'] = dg['p-value'].apply(binarize) # 1 if significant 

dg.columns = [' '.join(i.split('.')) for i in dg.columns.tolist()] # format columns
```

```python
keys = dg['GLM Parameter'].unique().tolist()
values_GLM, values_Q = [], []
for k in keys:
    if 'Effect Size' in k:
        values_GLM += ['Slope']
    else:
        values_GLM += ['Intercept']
    if 'Sample Quality' in k:
        values_Q  += ['High Quality']
    else:
        values_Q += ['Low Quality']
    
dg['GLM Interpretation'] = dg['GLM Parameter'].map(dict(zip(keys,values_GLM)))
dg['Quality Interpretation'] = dg['GLM Parameter'].map(dict(zip(keys,values_Q)))
```

```python
sns.set(style = 'white', font_scale=2)
quality = ['Avg Freeze Thaw', 'Avg RIN']

fig, ax = plt.subplots(ncols = len(quality), figsize = (16,6))

for j in range(len(quality)):
    q = quality[j] # for i in range(len(quality)):
    dg_q = dg[dg['Sample Quality Metric'] == q]

    a = np.char.array(dg_q['Subset Size'].values)
    b = np.char.array(dg_q['GLM Interpretation'].values)
    dg_q['X axis'] = (a + b'_' + b).astype(str)
    dg_q.sort_values(by = ['Subset Size', 'GLM Interpretation', 'Quality Interpretation'], inplace = True)
    dg_q.reset_index(inplace = True, drop = True)
    dg_q.drop(index = dg_q[dg_q['GLM Interpretation'] == 'Intercept'].index, inplace = True)
    dg_q.reset_index(inplace = True, drop = True)

    blue = sns.color_palette("RdBu_r", 7)[0]
    red = sns.color_palette("RdBu_r", 7)[-1]
    colors = [blue, red]
    blue_ = sns.color_palette("RdBu_r", 2)[0]
    red_ = sns.color_palette("RdBu_r", 5)[-2]

    counter = 0
    if len(dg_q['p-value'].unique()) == 2:
        g = sns.scatterplot(x = 'X axis', y = 'Estimate', data = dg_q, 
                        hue = 'Quality Interpretation', palette = colors, style = 'p-value', s = 70, zorder = 2, 
                            style_order = [1,0],
                        ax = ax[j])
    #         legend = ax[j].legend()
        ax[j].legend().set_visible(False)
        leg_ind = j
        counter += 1
    else:
        g = sns.scatterplot(x = 'X axis', y = 'Estimate', data = dg_q, 
                        hue = 'Quality Interpretation', palette = colors, s = 70, zorder = 2, ax = ax[j])
        ax[j].legend().set_visible(False)
        leg_ind_ = j

    x_ = ax[j].get_xticks() 
    ymin_good, ymin_bad, ymax_good, ymax_bad, x_good, x_bad = [], [], [],[],[],[]
    for k in range(0,dg_q.shape[0],2):
        if dg_q.loc[k, 'Estimate'] < dg_q.loc[k + 1, 'Estimate']:
            ymin_good += [dg_q.loc[k, 'Estimate']]
            ymax_good += [dg_q.loc[k+1, 'Estimate']]
            x_good += [x_[int(k/2)]]
        else:
            ymin_bad += [dg_q.loc[k, 'Estimate']]
            ymax_bad += [dg_q.loc[k+1, 'Estimate']]
            x_bad += [x_[int(k/2)]]


    g.vlines(x = x_good, ymin = ymin_good, ymax = ymax_good, colors = blue_, linestyles='--', zorder = 1)
    g.vlines(x = x_bad, ymin = ymin_bad, ymax = ymax_bad, colors = red_, linestyles='--', zorder = 1)

    fig.canvas.draw()
    labels = [item.get_text() for item in ax[j].get_xticklabels()]
    subset_size_label = list(map(str,sorted(set([int(k.split('_')[0]) for k in labels]))))
    ax[j].set_xticklabels(subset_size_label)
    ax[j].set_ylabel('\u0394D')
    ax[j].set_xlabel('Subset Size')
    ax[j].yaxis.set_major_locator(ticker.MultipleLocator(0.1))
    ax[j].set_ylim([-0.08,0.78])
#     ax[j].tick_params(axis='x', which='major', labelsize=1)
ax[0].set_title('Freeze-Thaw')
ax[1].set_title('RIN')
ax[1].set_ylabel('')
ax[1].get_yaxis().set_ticks([])
fig.subplots_adjust(wspace = 0.1)

fig.show();
```

```python
cg = pd.read_csv('../data/supplementary_table_similarity_2.csv')
cg = cg[(cg['Normalization Method'] == normalization_type)].iloc[:,1:]
cg.drop(index = cg[cg['Estimate'].isna()].index, inplace = True)
cg['p-value'] = cg['p-value'].apply(binarize)
cg.reset_index(inplace = True, drop = True)
```

```python
glms = cg['GLM'].unique().tolist()
counter = 0
# colors = colors = sns.color_palette("husl")[:len(vals)]
sns.set(style = 'white', font_scale=1.42)
fig, ax = plt.subplots(ncols = len(glms), figsize = (16,5))
for j in range(len(glms)):

    cg_g = cg[cg['GLM'] == glms[j]]
    cg_g['X axis']= range(cg_g.shape[0])
    cg_g['X axis'] = cg_g['X axis'].astype(str)
    
    if j == 0:
        vals = cg_g['GLM Variables'].tolist()
        colors = sns.color_palette("husl")[:len(vals)]
        cmap = dict(zip(vals, colors))
    else:
        vals = cg_g['GLM Variables'].tolist()
        colors = [cmap[v] for v in vals]
    
    cg_g.reset_index(inplace = True, drop = True)
    

    if len(cg_g['p-value'].unique()) == 2:
        g = sns.scatterplot(x = 'X axis', y = 'Estimate', data = cg_g, hue = 'GLM Variables', zorder = 2, ax = ax[j],
                            palette = colors,style = 'p-value', style_order = [1,0], s = 70)
        ax[j].legend().set_visible(False)
        leg_ind = j
        counter += 1
    else:
        g = sns.scatterplot(x = 'X axis', y = 'Estimate', data = cg_g, hue = 'GLM Variables', zorder = 2, ax = ax[j],
                           palette = colors, s = 70)
        ax[j].legend().set_visible(False)


    x_ = ax[j].get_xticks()
    ymin_, ymax_ = [], []
    for k in cg_g.index:
        val = cg_g.loc[k, 'Estimate']
        if  val < 0:
            ymin_ += [val]
            ymax_ += [0]
        else:
            ymin_ += [0]
            ymax_ += [val]
    g.vlines(x = cg_g.index, ymin = ymin_, ymax = ymax_, colors = 'grey', linestyles='--', zorder = 1)
    ax[j].axhline(0, c = 'black', lw = 1, xmin = 0, xmax = 3)
#     ax[j].set_xlabel(glms[j])
    ax[j].set_xlabel('')
    ax[j].set_ylabel('')
    if j > 0:
        ax[j].set_xticklabels(cg_g['GLM Variables'].tolist(), rotation = 0, fontsize = 23)
    else: 
        ax[j].set_xticklabels(cg_g['GLM Variables'].tolist(), rotation = 45, fontsize = 23)
    labels = [item.get_text() for item in ax[j].get_xticklabels()]
    labels[0] = '$\mathregular{S_0}$'
    ax[j].set_xticklabels(labels)

    ylim_ = ax[j].get_ylim()
    interval = 0.05
    
    def myround(x, prec=2, base=.05):
        return round(base * round(float(x)/base),prec)
    ax[j].yaxis.set_ticks(np.arange(myround(ylim_[0]), myround(ylim_[1] + interval), interval))
    
    if j > 0:
        xlim_ = ax[j].get_xlim()
        window = 0.25
        ax[j].set_xlim(xlim_[0] - window, xlim_[1] + window)

    ax[j].set_xlabel('')
    ax[j].set_ylabel('Similarity', size = 20)
    ax[j].tick_params(axis='x', which='major', labelsize=20)
    ax[j].tick_params(axis='y', which='major', labelsize=20)

# for i in range(len(labels)):
#     if i % 2 == 0:
#         labels[i] = ''
# ax[2].set_yticklabels(labels) 
ax[0].yaxis.set_ticks(np.arange(-0.4,0.11, 0.1))
ax[2].yaxis.set_ticks(np.arange(-0.5,0.2, 0.1))
    
handles_, labels_ = ax[0].get_legend_handles_labels()
handles_, labels_ = handles_[1:], labels_[1:]

if counter > 0:
    handles, labels = ax[leg_ind].get_legend_handles_labels()
    handles, labels = handles[-2:], labels[-2:]
    first_p = cg[cg['GLM'] == glms[j]].iloc[0, cg.columns.tolist().index('p-value')]
    labels[1] = 'p > 0.05'
    labels[0] = 'p $\leq$ 0.05'
    
    handles_ += handles
    labels_ += labels

# print(colors)
fig.subplots_adjust(wspace = 0.38)
fig.legend(handles_, labels_, loc = 'upper right', bbox_to_anchor = (0.07,-0.1,1.01,1),fontsize=20,
            bbox_transform = plt.gcf().transFigure)
```

```python
def binarize(x):
    if x > 0.05:
        return 0
    else:
        return 1
```

```python
dg = pd.read_csv('../data/supplementary_noise_glms.csv', index_col = 0)
dg.drop(columns = ['SE','t'], inplace = True)
dg.rename(columns = {'Pr_t': 'p-value'}, inplace = True)
dg['p-value'] = dg['p-value'].apply(binarize)
dg.reset_index(inplace = True, drop = True)

dg.Distance_Metric = dg.Distance_Metric.astype("category")
dg.Distance_Metric.cat.set_categories(['euclidean', 'rmse', 'pearson', 'spearman'], inplace=True)
dg.sort_values(["Distance_Metric"], inplace = True)
```

```python
sns.set(style = 'white', font_scale=1.3)
fig, ax = plt.subplots(ncols = 2, figsize = (10,5))

for j in range(0,2):
    if j == 0:
        dg_f = dg[dg['GLM']=='(Noise ~ Freeze Thaw)']
        dg_f.sort_values(by = ['Distance_Metric', 'Parameter'], inplace = True)
        dg_f['Parameter'] = dg_f['Parameter'].map({'Base Noise': 'Intercept', 'Freeze Thaw Coefficient': 'Slope'})
        colors = sns.color_palette("husl")[:3]
        colors = [colors[0]] + [colors[2]]
        # colors = colors*4
        a = np.char.array(dg_f['Distance_Metric'].values)
        b = np.char.array(dg_f['Parameter'].values)
        dg_f['X axis'] = (a + b'_' + b).astype(str)

        g = sns.scatterplot(x = 'X axis', y = 'Estimate', data = dg_f, style = 'p-value', s = 70, zorder = 2, 
                                style_order = [1,0], ax = ax[j], hue = 'Parameter', palette = colors)


        x_ = ax[j].get_xticks()
        ymin_, ymax_ = [], []
        for k in dg_f.index:
            val = dg_f.loc[k, 'Estimate']
            if  val < 0:
                ymin_ += [val]
                ymax_ += [0]
            else:
                ymin_ += [0]
                ymax_ += [val]
        g.vlines(x = dg_f.index, ymin = ymin_, ymax = ymax_, colors = 'grey', linestyles='--', zorder = 1)
        ax[j].axhline(0, c = 'black', lw = 1, xmin = 0, xmax = 3)
        ax[j].set_xlabel('GLM(Noise ~ Freeze Thaw)')
#         ax[j].legend().set_visible(False)
        handles, labels = ax[j].get_legend_handles_labels()
        handles, labels = handles[1:], labels[1:]
        labels[-1] = 'p > 0.05'
        labels[-2] = 'p $\leq$ 0.05'
        ax[0].legend(handles, labels, bbox_to_anchor = [-0.2,1])

    if j == 1:
        dg_f = dg[(dg['GLM']=='(Noise ~ RIN), 1 Freeze Thaw') | (dg['GLM']=='(Noise ~ RIN), 2 Freeze Thaws')]
        dg_f['Parameter'] = dg_f['Parameter'].map({'Base Noise': 'Intercept', 'RIN Coefficient': 'Slope'})
        dg_f['Freeze Thaw'] = dg_f['GLM'].apply(lambda x: int(x.split('),')[1][1]))
        dg_f.sort_values(by = ['Distance_Metric', 'Parameter', 'Freeze Thaw'], inplace = True)

        a = np.char.array(dg_f['Distance_Metric'].values)
        b = np.char.array(dg_f['Parameter'].values)
        dg_f['X axis'] = (a + b'_' + b).astype(str)

        dg_f.reset_index(inplace = True, drop = True)

        blue = sns.color_palette("RdBu_r", 7)[0]
        red = sns.color_palette("RdBu_r", 7)[-1]
        colors = [blue, red]
        blue_ = sns.color_palette("RdBu_r", 2)[0]
        red_ = sns.color_palette("RdBu_r", 5)[-2]


        g = sns.scatterplot(x = 'X axis', y = 'Estimate', data = dg_f, style = 'p-value', s = 70, zorder = 2, 
                                style_order = [1,0], ax = ax[j], hue = 'Freeze Thaw', palette = colors)


        x_ = ax[j].get_xticks() 
        ymin_good, ymin_bad, ymax_good, ymax_bad, x_good, x_bad = [], [], [],[],[],[]
        for k in range(0,dg_f.shape[0],2):
            if dg_f.loc[k, 'Estimate'] < dg_f.loc[k + 1, 'Estimate']:
                ymin_good += [dg_f.loc[k, 'Estimate']]
                ymax_good += [dg_f.loc[k+1, 'Estimate']]
                x_good += [x_[int(k/2)]]
            else:
                ymin_bad += [dg_f.loc[k, 'Estimate']]
                ymax_bad += [dg_f.loc[k+1, 'Estimate']]
                x_bad += [x_[int(k/2)]]


        g.vlines(x = x_good, ymin = ymin_good, ymax = ymax_good, colors = blue_, linestyles='--', zorder = 1)
        g.vlines(x = x_bad, ymin = ymin_bad, ymax = ymax_bad, colors = red_, linestyles='--', zorder = 1)
        ax[j].set_xlabel('GLM(Noise ~ RIN)')
        ax[j].set_ylabel('')
        ax[j].legend().set_visible(False)

    fig.canvas.draw()
    labels = [item.get_text() for item in ax[j].get_xticklabels()]
    subset_size_label = [k.split('_')[0] for k in labels][1::2]
    
    labels = [k.split('_')[1] for k in labels]
    nl = {'Intercept': '$\mathregular{N_0}$', 'Slope': '\u0394N'}
    labels = [nl[v] for v in labels]
    ax[j].set_xticklabels(labels, rotation = 45)
    
    # second layer of labels
    # "Activate" minor ticks
    ax[j].minorticks_on()

    # Get location of the center
    locs = []
    for k in range(0,len(ax[j].get_xticks()),2):
        locs += [(x_[k] + x_[k + 1])/2]
    # Set minor ticks there
    ax[j].set_xticks(locs, minor = True)

    # Set the labels
    from matplotlib import ticker
    ax[j].xaxis.set_minor_formatter(ticker.FixedFormatter(subset_size_label))  #add the custom ticks

    # Move the category label further from x-axis
    ax[j].tick_params(axis='x', which='minor', pad=60)

    # Remove minor ticks where not necessary
    ax[j].tick_params(axis='x',which='both', top=False)
    ax[j].tick_params(axis='x',which='minor', bottom=False)
    ax[j].tick_params(axis='y',which='both', left=False, right = False)
    
    
  
    
handles, labels = ax[1].get_legend_handles_labels()
labels[-1] = 'p > 0.05'
labels[-2] = 'p $\leq$ 0.05'
ax[0].set_ylabel('Noise')
# print(colors)
# fig.subplots_adjust(wspace = 0.3)
fig.legend(handles, labels, loc = 'upper right', bbox_to_anchor = (0.12,-0.12,1,1),
            bbox_transform = plt.gcf().transFigure)
```
