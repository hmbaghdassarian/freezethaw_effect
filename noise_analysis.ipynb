```r
### functions
g_legend <- function(a.gplot){
  tmp <- ggplot_gtable(ggplot_build(a.gplot))
  leg <- which(sapply(tmp$grobs, function(x) x$name) == "guide-box")
  legend <- tmp$grobs[[leg]]
  return(legend)}
                      
univar_mod<-function(vars,df,keep,context_var,resp){
    mods=as.data.frame(do.call(rbind, lapply(vars, function(s){ 
        cbind(as.data.frame(coef(summary(glm(formula(paste(resp,'~',context_var,'+',s)), data=df[keep,] )))),
              cbx=s)
    })))
    colnames(mods) = c('Estimate','SE','t','Pr_t','cbx')
    mods$Var = as.character(gsub('[1-9]','',rownames(mods)))
    mods$Var = gsub('\\(Intercept)','Base MCN',mods$Var)
    mods$Var = gsub('concentration','MCN Increase / (ng/uL)',mods$Var)
    mods$Var[grepl('cbx_',mods$Var)] = 'Var'
    # visualize GLMs
    limits <- aes(ymax = Estimate + SE*1.96, ymin=Estimate - SE*1.96)
    #head(mods)
    g=ggplot(mods, aes( y=Estimate, x=Var,colour=-log(Pr_t,10))) +
        geom_point() + geom_errorbar(limits, width=0.1) + #scale_color_continuous(high='red',low='blue') +
        scale_color_gradientn(colors = rainbow(10))+
        facet_grid(~cbx) + #theme(axis.text.x = element_text(angle = 60, hjust = 1)) +
        ylab('Similarity') + geom_hline(yintercept = 0)+ theme_classic()+ theme(axis.text.x = element_text(angle = 60, hjust = 1))
    return(g)
}
```

## Load sample annotation
Sample annotation includes diagnosis, total input RNA, concentration, number of freeze-thaw cycles

```r
annot = read.xlsx('manifest.xlsx',sheet = 6)
all(annot$DX == annot$DX_check)
annot$DX = factor( ifelse(grepl('ASD',annot$DX),'ASD',ifelse(grepl('TD',annot$DX),'TD',NA)) , levels = c('TD','ASD'))
annot$concentration_bins = cut(annot$concentration,breaks=seq(15,105,15))
annot$final_freeze_thaw_fact = factor(as.character(annot$final_freeze_thaw),levels=c('1','2'))

annot$SampleName = as.factor(annot$SampleName)
annot$Gender = as.factor(annot$Gender)

blooddata = unique(read.csv( 'data/polyA-data_updated.csv')[,c('blooddrawId','subjectid','ageMo','collection_year')])
annot = merge(annot,blooddata,all.x=T,by='blooddrawId')
annot = annot[order(annot$Order),]

# load data
l=list()
ls = list.files('htseq/',pattern = '*.txt')
for(i in ls){
    l[[i]] = read.csv(paste0('htseq/',i),sep='\t',header=F,row.names=1)
}

# format expression
expss = do.call(cbind,l)
colnames(expss) = unlist(lapply( strsplit(names(l),'_') , function(x) x[2] ))
expss = expss[,order(as.numeric(gsub('S','',colnames(expss))))]
expss = expss[(rowMeans(expss[,which(annot$DX=='ASD')])>5|rowMeans(expss[,which(annot$DX=='TD')])>5) & rowMeans(expss)<( quantile(rowMeans(expss),c(.10,.50,.90,.95,.97,.999))[6] ) & grepl('ENSG.*',rownames(expss)),]
rm = c('S0') 
expss = expss[,!(colnames(expss)%in%rm)]
counts = expss
cpm = apply(counts,2,function(x) (x/sum(x))*1e6)
expss = cpm
            
# match annot and expss
annot = annot[annot$Comment %in% colnames(expss),]
rownames(annot)=annot$Comment  
            
# gene length
if(F){
    genelength = unlist( lapply(genes,function(x) sum(ranges(union(x,x))@width)) )
    save(unlist(genelength),file='gene_length.grch38.rda')
}else{
    load('gene_length.grch38.rda')
    genelength = sapply(rownames(expss),function(x) genelength[strsplit(x,'\\.')[[1]][1]])
}

RPKM <- function(count_i,gene_length) (count_i/((sum(count_i)/1e6)*(gene_length/1e3)))
```

# Calculate replicate noise

```r
replicates = unique( annot$SampleName[annot$Sample_type!='Industry'])
replicate_expss_sim = list()
metrics = c('euclidean','rmse','pearson','spearman') 
p=seq(4,20,.5)*.01

for(rep in replicates){
    dat_counts = expss[,annot$SampleName==rep]
    dat = induced_noise_controls(dat_counts,p=p,assumption = 'poisson')
    rpkm = apply(dat,2,function(x) RPKM(count_i = x,gene_length = genelength))
    logdat = log(rpkm+1)
    tmp_i = do.call(rbind,lapply(metrics,function(meth){
        similarity(logdat,method=meth,rep=rep,plot_on=plot_on)
        }))
    replicate_expss_sim[[rep]] = tmp_i
    }

# further sample descriptoin: length and GC bias (cor between count and genelength or bias), base stats, median, quartiles, variance
flatness = function(x,n=100){ mean( sapply(1:n,function(y) mean( abs(sample(x)-sample(x)) ) )) }
tmp_sample_stats = data.frame(mean=apply(expss,2,mean),var=apply(expss,2,var),
                              quantile=as.data.frame( t(apply(expss,2,quantile)) ),
                              flatness=apply(expss,2,flatness),norm=apply(expss,2,function(x) norm(x,type='2')),
                                  library_size=apply(expss,2,sum))
colnames(tmp_sample_stats) = paste('Parent',colnames(tmp_sample_stats),sep='_')
tmp_sample_stats$Parent_Sample = rownames(tmp_sample_stats)
                                  
tmp_orig_sample_stats = tmp_sample_stats
colnames(tmp_orig_sample_stats) = gsub('Parent','Original',colnames(tmp_orig_sample_stats))
                                  
sample_stats = c( colnames(tmp_sample_stats),colnames(tmp_orig_sample_stats))
```

```r
g_out = lapply(replicate_expss_sim,function(tmp_i){
    tmp_i = tmp_i %>% group_by(Similarity_Metric) %>% mutate(min_cross_noise_mean = mean(min_cross_noise))
    tmp_i = tmp_i %>% group_by(Similarity_Metric) %>% mutate(min_cross_sim_mean = mean(min_cross_sim))
    g=ggplot(data=tmp_i,aes(x=Percent_Noise,y=Similarity,color=self_or_replicate))+geom_point(size=1.2)+
    scale_color_manual('',breaks = c("Original", "Replicate"), labels = c('Reference', 'Target'), 
                       values=c("dodgerblue3", "brown2"))+
  theme(text = element_text(size=24), panel.background = element_blank(), 
        panel.border = element_rect(colour = 'grey', fill = 'NA', size = 1.2),
       panel.grid.major = element_line(colour = 'grey', size = 0.3))+
        facet_grid(Similarity_Metric~.,scales='free')+stat_smooth() + ylab('Dissimilarity') + xlab('Noise')+
        geom_segment(aes(yend = min_cross_sim_mean,y= first_replicate,x=0,xend=min_cross_noise_mean),size=1,col='black')
    g
})

g_out[[1]]
```
