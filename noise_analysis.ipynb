```r
### functions
g_legend <- function(a.gplot){
  tmp <- ggplot_gtable(ggplot_build(a.gplot))
  leg <- which(sapply(tmp$grobs, function(x) x$name) == "guide-box")
  legend <- tmp$grobs[[leg]]
  return(legend)}
                      
univar_mod<-function(vars,df,keep,context_var,resp){
    mods=as.data.frame(do.call(rbind, lapply(vars, function(s){ 
        cbind(as.data.frame(coef(summary(glm(formula(paste(resp,'~',context_var,'+',s)), data=df[keep,] )))),
              cbx=s)
    })))
    colnames(mods) = c('Estimate','SE','t','Pr_t','cbx')
    mods$Var = as.character(gsub('[1-9]','',rownames(mods)))
    mods$Var = gsub('\\(Intercept)','Base MCN',mods$Var)
    mods$Var = gsub('concentration','MCN Increase / (ng/uL)',mods$Var)
    mods$Var[grepl('cbx_',mods$Var)] = 'Var'
    # visualize GLMs
    limits <- aes(ymax = Estimate + SE*1.96, ymin=Estimate - SE*1.96)
    #head(mods)
    g=ggplot(mods, aes( y=Estimate, x=Var,colour=-log(Pr_t,10))) +
        geom_point() + geom_errorbar(limits, width=0.1) + #scale_color_continuous(high='red',low='blue') +
        scale_color_gradientn(colors = rainbow(10))+
        facet_grid(~cbx) + #theme(axis.text.x = element_text(angle = 60, hjust = 1)) +
        ylab('Similarity') + geom_hline(yintercept = 0)+ theme_classic()+ theme(axis.text.x = element_text(angle = 60, hjust = 1))
    return(g)
}
```

# 1. Make Measurements of Replicability and Bias

In this section we will examine the count and DE similarity within technical and industry replicates as measured by correlation and RSME. GC-bias will be measured by association between counts, DE stats and GC content


##  1.1 Measure correlation/RSME of counts within technical replicates and quality matched industry standards
Measures of replicability will be made between technical replicates to assess the consistency of measurements resulting from method and measurement error. Error from measurement and method will not be dissociated in this examination.

### 1.1.1 Consistency of Technical Sample Replicates
Here we measure the consistency of counts between technical sample replicates.



```r
par=T
use_cmp = TRUE
plot_on=TRUE

replicates = unique( annot$SampleName[annot$Sample_type!='Industry'])

#if(par){ registerDoSNOW(makeCluster(6, type = "SOCK")) }
if(par){registerDoMC(cores=min(length(replicates),25))}

# technical replicates
replicate_expss_sim = list()
g_out=list()
min=4
 max=20
p=seq(min,max,.5)*.01
metrics = c('euclidean','rmse','pearson','spearman') #,'poissonProb') 
if(par){
    replicate_expss_sim = foreach(rep=replicates,.packages = 'reshape') %dopar% {
        print(rep)
        dat_counts = expss[,annot$SampleName==rep]
        dat = induced_noise_controls(dat_counts,p=p,assumption = 'poisson')
        #cpm = apply(dat,2,function(x) (x/sum(x))*1e6) # convert to cpm 
        rpkm = apply(dat,2,function(x) RPKM(count_i = x,gene_length = genelength))
        logdat = log(rpkm+1)
        tmp_i = do.call(rbind,lapply(metrics,function(meth){
            if(meth=='poissonProb'){
                similarity(dat,method=meth,rep=rep,plot_on=plot_on)
            }else{
                similarity(logdat,method=meth,rep=rep,plot_on=plot_on)
            }
            }))
        tmp_i   
    }
}else{
    for(rep in replicates){
        print(rep)
        dat_counts = expss[,annot$SampleName==rep]
        dat = induced_noise_controls(dat_counts,p=p,assumption = 'poisson')
        #cpm = apply(dat,2,function(x) (x/sum(x))*1e6) # convert to cpm 
        rpkm = apply(dat,2,function(x) RPKM(count_i = x,gene_length = genelength))
        logdat = log(rpkm+1)
        tmp_i = do.call(rbind,lapply(metrics,function(meth){
            if(meth=='poissonProb'){
                similarity(dat,method=meth,rep=rep,plot_on=plot_on)
            }else{
                similarity(logdat,method=meth,rep=rep,plot_on=plot_on)
            }
            }))
        replicate_expss_sim[[rep]] = tmp_i 
    } 
}
                    
```

```r
# further sample descriptoin
# length and GC bias (cor between count and genelength or bias)
# base stats, median, quartiles, variance
flatness = function(x,n=100){ mean( sapply(1:n,function(y) mean( abs(sample(x)-sample(x)) ) )) }
tmp_sample_stats = data.frame(mean=apply(expss,2,mean),var=apply(expss,2,var),
                              quantile=as.data.frame( t(apply(expss,2,quantile)) ),
                              flatness=apply(expss,2,flatness),norm=apply(expss,2,function(x) norm(x,type='2')),
                                  library_size=apply(expss,2,sum))
colnames(tmp_sample_stats) = paste('Parent',colnames(tmp_sample_stats),sep='_')
tmp_sample_stats$Parent_Sample = rownames(tmp_sample_stats)
                                  
tmp_orig_sample_stats = tmp_sample_stats
colnames(tmp_orig_sample_stats) = gsub('Parent','Original',colnames(tmp_orig_sample_stats))
                                  
sample_stats = c( colnames(tmp_sample_stats),colnames(tmp_orig_sample_stats))
```
